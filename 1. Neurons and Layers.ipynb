```python
import numpy as np
import matplotlib.pyplot as plt 
import tensorflow as tf
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras import Sequential
from tensorflow.keras.losses import MeanSquaredError, BinaryCrossentropy
```

```python
X_train = np.array([[1.0],[2.0]], dtype=np.float32)
Y_train = np.array([[300.0],[500.0]], dtype=np.float32)

# fig, ax = plt.subplots(1,1)
# ax.scatter(X_train,Y_train,marker='x',c='r',label="Data Points")
# ax.legend(fontsize='xx-large')
# ax.set_ylabel('Price (in 1000s of dollars)', fontsize = 'xx-large')
# ax.set_xlabel('Size (1000 sqft)', fontsize ='xx-large')
# plt.show()
print(X_train)
print(Y_train)
```

## Regression model

$$ f_{\mathbf{w},b}(x^{(i)}) = \mathbf{w}\cdot x^{(i)} + b \tag{1}$$

```python
linear_layer = Dense(units=1, activation='linear')
```

```python
a1=linear_layer(X_train[0].reshape(1,1))
print(a1)
```

```python
w,b=linear_layer.get_weights()
print(f"w: {w}, b: {b}")
```

```python
set_w = np.array([[200]])
set_b = np.array([100])

linear_layer.set_weights([set_w,set_b])
print(linear_layer.get_weights())
```

```python
a1 = linear_layer(X_train[0].reshape(1,1))
a2 = linear_layer(X_train[1].reshape(1,1))

print(a1)
print(a2)
```

```python
print(np.dot(set_w,X_train[0].reshape(1,1))+set_b)
print(np.dot(set_w,X_train[1].reshape(1,1))+set_b)
```

```python
prediction = linear_layer(X_train)
prediction_np = np.dot(X_train,set_w)+set_b
```

```python
plt.plot(X_train,prediction,'r', marker='x')
plt.title('np prediction')
plt.ylabel('Price in (1000s) of dollars')
plt.xlabel('Size (1000 sqft)')
```

```python
plt.plot(X_train,prediction_np,'b',marker='o')
plt.title('np prediction')
plt.ylabel('Price in (1000s) of dollars')
plt.xlabel('Size (1000 sqft)')
```
